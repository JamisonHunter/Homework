# Describing Distributions

```python
# Imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# setting the style for seaborn graphs
plt.style.use(['dark_background','seaborn-muted', 'seaborn-poster'])
```

```python
# gathering file path to transform into a data frame
path = r"C:\Users\Jamison\Desktop\CODINGDOJO\data\data.csv"
df = pd.read_csv(path)
df.head()
```

```python
mean = df['VitD_levels'].mean()
std = df['VitD_levels'].std()
median = df['VitD_levels'].median()
plus_one_std = mean + std
minus_one_std = mean - std

ax = sns.histplot(data=df, x='VitD_levels', discrete=True, stat='density', kde = True);

ax.axvline(mean, color = "red", label = "Mean");
ax.axvline(median, ls='--', color = "green", label = "Median");
ax.axvline(plus_one_std, color = "black");
ax.axvline(minus_one_std, color = "black");
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)

ax.set_ylabel("Density")
ax.set_xlabel("Vitamin D Levels")
ax.set_title("Vitamin D Distribution")
ax.legend()
```

```python
kurt = stats.kurtosis(df['VitD_levels'], fisher = False)
print(kurt)
if round(kurt) > 3:
    print("Leptokurtic")
if round(kurt) < 3:
    print("Platykurtic")
if round(kurt) == 3:
    print("Mesokurtic")
```

```python
print("The above graph is discrete with essentially no skew.")
```
